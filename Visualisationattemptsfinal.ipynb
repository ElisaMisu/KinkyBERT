```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/subreddits23/adjusted2_topics_info.csv')
```

```python
# Exclude the topic '-1'
df = df[df['Topic'] != -1]

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for the plots
sns.set(style="whitegrid")

# Plotting the top 10 topics by document count
top_topics = df.nlargest(25, 'Count')

plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(x='Count', y='Name', data=top_topics, palette='viridis')
bar_plot.set_title('Top 10 Topics by Document Count', fontsize=16)
bar_plot.set_xlabel('Document Count', fontsize=14)
bar_plot.set_ylabel('Topic Name', fontsize=14)
bar_plot.set_yticklabels(bar_plot.get_yticklabels(), rotation=0, fontsize=12)
plt.show()
```

```python
# Calculating the distribution of topics
topic_distribution = df['Name'].value_counts()

plt.figure(figsize=(10, 10))
pie_chart = plt.pie(topic_distribution, labels=topic_distribution.index, autopct='%1.1f%%', colors=sns.color_palette('pastel')[0:len(topic_distribution)])
plt.title('Distribution of Topics', fontsize=10)
plt.show()
```
